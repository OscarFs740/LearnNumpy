```python
from matplotlib import pyplot as plt
import random


x = range(0,120)
y = [random.randint(20,35) for i in range(120)]


#设置图片大小
plt.figure(figsize=(25,10),dpi=100)    #长  高   dpi


#绘图
plt.plot(x,y)

#设置轴的刻度
_xtick_labels = [i for i in range(130)]     #设置刻度list
plt.xticks(_xtick_labels[::10])             #取刻度

plt.yticks(range(min(y),max(y)+1))


#调整X轴的刻度
  


plt.show()

print([i for i in range(60)])
```
